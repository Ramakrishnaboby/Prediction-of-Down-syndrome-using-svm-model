```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
from sklearn.metrics import confusion_matrix 
from sklearn.model_selection import train_test_split
```

```python
#Shuffling the data for better modelling
df= pd.read_excel('C:\\Users\\admin\\OneDrive\\Desktop\\ML A2\\Data_Cortex_Nuclear.xlsx')
df = df.sample(frac=1).reset_index(drop=True)
df
```

```python
#Creating x and y
x = df.iloc[:, 1:-1]
y = df.iloc[:, -1].values
#Label encoding the last column of strings to integers
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
y = le.fit_transform(y)
y
```

```python
#Filling in the missing values
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer

# Create an IterativeImputer object
imputer = IterativeImputer(random_state=0)

# Fit the imputer to your data
x = pd.DataFrame(imputer.fit_transform(x))
x
```

```python
#Converting the values to the same scale
from sklearn.preprocessing import StandardScaler
ss = StandardScaler()
x = pd.DataFrame(ss.fit_transform(x))
x
```

```python
#Creating train data and test data
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size= 0.2)
```

```python
#Fitting the data into the SVM model
from sklearn.svm import SVC
classifier = SVC (kernel = 'rbf')
classifier.fit (x_train,y_train)
```

```python
#The predicted data for the test data
y_pred = classifier.predict(x_test)
y_pred
```

```python
#The predicted values by the model for the entire data
Y_pred= classifier.predict(x)
print (Y_pred)

#The predicted data for the entire model
df_ypred = pd.DataFrame(Y_pred,columns = ['Prediction'])
result = pd.DataFrame(pd.concat ([x,pd.DataFrame(y),df_ypred],axis = 1 ))
print(result)
```

```python
#The accuracy score
from sklearn.metrics import accuracy_score
print (accuracy_score (y_test,y_pred))
```

```python
#The confusion matrix showing the correctly predicted and incorrectly predicted amount
cm = confusion_matrix (y_test,y_pred)
print (cm)
```
